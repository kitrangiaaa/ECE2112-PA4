__a. Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as Instrumentation and hometown Luzon__

__b. Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female__

```python
boards['Average'] = boards[['Math','Electronics','GEAS','Communication']].mean(axis=1)
boards
```

### __ECE BOARD EXAM PROBLEM 2__
#### __2. Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown contributes to a higher average score?__

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
plt.bar(boards['Track'], boards['Average'], color='#B7D0CD')
plt.xlabel("Track")
plt.ylabel("Average Score")
plt.title("Correlation of the Chosen Track and Average of Students", fontsize=15)

plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(boards['Gender'], boards['Average'], color='#E7CDD6')
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.title("Correlation of the Gender and Average of Students", fontsize=15)

plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(boards['Hometown'], boards['Average'], color='#D5CBF0')
plt.xlabel("Hometown")
plt.ylabel("Average Score")
plt.title("Correlation of the Hometown and Average of Students", fontsize=15)

plt.show()
```
